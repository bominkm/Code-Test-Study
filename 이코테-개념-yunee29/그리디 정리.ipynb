## (1) 최적의 방법 찾는 알고리즘

```python
# 문자열 뒤집기
data = input()
count0 = 0
count1 = 0

if data[0] == '1':
  count0 += 1
else:
  count1 += 1

for i in range(len(data)-1):
  if data[i] != data[i+1]:
    if data[i+1] == '1':
      count0 += 1
    else:
      count1 += 1
```

## (2) sort시킨 후 풀기

```python
# 만들 수 없는 금액
n = int(input())
data = list(map(int, input().split()))
data.sort()

target = 1
for x in data:
  if target < x:
    break
  target += x

print(target)
```
