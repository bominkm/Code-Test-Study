모험가 길드

```python
n = int(input())
data = list(map(int, input().split()))
data.sort()

result = 0
count = 0

for i in data: # 1 2 2 2 3
  count += 1
  if count >= i: 
    result += 1
    count = 0 
```

곱하기 혹은 더하기

```python
data = input()

result = int(data[0])

for i in range(1, len(data)):
  num = int(data[i])
  if num <= 1 or result <= 1:
    result += num
  else:
    result *= num
```

문자열 뒤집기

```python
data = input()
count0 = 0
count1 = 0

if data[0] == '1':
  count0 += 1
else:
  count1 += 1

for i in range(len(data)-1):
  if data[i] != data[i+1]:
    if data[i+1] == '1':
      count0 += 1
    else:
      count1 += 1
```

만들 수 없는 금액

```python
n = int(input())
data = list(map(int, input().split()))
data.sort()

target = 1
for x in data:
  if target < x:
    break
  target += x

print(target)
```

연습

```python
n = int(input())
scar = list(map(int, input().split()))
scar.sort()

group = 0
count = 0

for s in scar:
  count += 1
  if count >= s:
    group += 1
    count = 0

print(group)
```

```python
num = input()

result = int(num[0])

for i in range(1, len(num)):
  if int(num[i]) <= 1 or result <= 1:
    result += int(num[i])
  else:
    result *= int(num[i])

print(result)
```

```python
string = input()
```

```python
count0 = 0
count1 = 0

if string[0] == 0:
  count1 += 1
else:
  count0 += 1

for i in range(len(string)-1):
  if string[i] != string[i+1]:
    if string[i+1] == '1':
      count0 += 1
    else:
      count1 += 1

print(min(count0, count1))
```
